```python
import re

def has_special_chars(input_string):
    """
    Checks if the input_string contains any character that is NOT
    a letter (a-z, A-Z) or a number (0-9).
    """
    # The pattern matches any character *not* in the specified range
    special_char_pattern = re.compile(r'[^a-zA-Z0-9]')

    # search() returns a match object if a match is found, otherwise None
    if special_char_pattern.search(input_string):
        return True
    else:
        return False
```

```python
columns_to_check = df.columns
has_special = df[columns_to_check].astype(str).apply(
    lambda x: x.str.contains(r'[^a-zA-Z0-9]', regex=True)
)

df['has_special_char'] = has_special.any(axis=1)
df['has_special_char']
```
