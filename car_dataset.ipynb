"Symboling" in an automotive context, specifically within datasets like the UCI Machine Learning Repository Automobile Data Set, refers to an insurance risk rating assigned to a vehicle. It indicates how much more or less risky a car is compared to its price, ranging from -3 (safe) to +3 (risky)

```python
df['company'] = df['car_name'].str.split().str[0]
df['company'] = df['company'].replace({
    'vokswagen': 'volkswagen',
    'volkswagon': 'volkswagen',
    'vw': 'volkswagen',
    'toyouta': 'toyota',
    'maxda': 'mazda',
    'Nissan' : 'nissan',
    'porcshce' : 'porsche',
})
company_counts = df['company'].value_counts()
```

```python
company_counts.plot(kind='bar', figsize=(12,6))
```

```python
ax = company_counts.plot(kind='bar', figsize=(12,6))

# Add numbers on top
for i, value in enumerate(company_counts):
    ax.text(i, value + 0.5, str(value), ha='center')

#If bars are tall, adjust spacing:
ax.set_ylim(0, company_counts.max() + 5)

plt.title("Number of Cars per Company")
plt.xlabel("Company")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
company_counts.sort_values(ascending=False).plot(kind='bar', figsize=(12,6))
plt.title("Number of Cars per Company")
plt.xlabel("Company")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
stack_data = pd.crosstab(df['company'], df['car_name'])
```

```python
stack_data.plot(
    kind='bar',
    stacked=True,
    figsize=(14,7),
    legend=False   # change to True if you want model legend
)

plt.title("Company-wise Car Model Distribution")
plt.xlabel("Company")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```
