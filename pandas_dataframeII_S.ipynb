# Merge the DataFrames on `student_id`

```python
# merge
df = df_students.merge(df_scores, on="student_id", how="inner")
df
```

```python
# merge with left
df_left = df_students.merge(df_scores, on="student_id", how="left")
df_left
```

```python
# merge with right
df_right = df_students.merge(df_scores, on="student_id", how="right")
df_right
```
