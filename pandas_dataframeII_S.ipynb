# Merge the DataFrames on `student_id`

```python
# merge
df = df_students.merge(df_scores, on="student_id", how="inner")
df
```

```python
#inner merge
df_inner = df_students.merge(df_scores, on="student_id", how="inner")
df_inner
```

```python
#outer merge
df_outer = df_students.merge(df_scores, on="student_id", how="outer")
df_outer
```

```python
# merge with left
df_left = df_students.merge(df_scores, on="student_id", how="left")
df_left
```

```python
# merge with right
df_right = df_students.merge(df_scores, on="student_id", how="right")
df_right
```

# Sorting

```python
# sort the values based on program
df.sort_values("program").head(10)
```

```python
# sort the values based on marks
df.sort_values("marks").head(10)
```

```python
# sort the values based on program and marks
df.sort_values(["program","marks"]).head(10) # sorts program 1st and if u have same name then goes for sorting on marks.
```

```python
df.sort_values(["marks","program"]).head(10) #sorts on marks if same number then goes for sorting on program.
```

```python
df.sort_values(["program","marks"], ascending=[True, False]).head(10) #same program name is printed on the index wise
```

```python
df.sort_values(["program","marks"], ascending=[ False, True]).head(10) #same marks are printed in the index order
```

```python
df.sort_values(["program","marks"], ascending=[ True, True]).head(10) #same program name is sorted on marks
```

```python
df.sort_values(["marks","program"], ascending=[ True, True]).head(10) #same marks number is sorted on program name
```

```python
df.sort_values(["program","marks"], ascending=[ False, False]).head(10) #its in decending order, same program name is in decreasing order of marks.
```
