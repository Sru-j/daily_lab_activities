#DATA CLEANING

SECTION A: COLUMN NAME CLEANING

```python
# Identify column names containing leading/trailing spaces
df.columns
```

```python
#Detect inconsistent capitalization in column headers.
df.columns = df.columns.str.strip().str.lower()
df.columns
```

```python
#Convert all column names to snake_case format.
df.columns = df.columns.str.replace(' ','_')
df.columns
```

```python
#Rename ‘irst Name’ to ‘First Name’.
df = df.rename(columns={'irst_name':'first_name'})
df.columns
```

 SECTION B: DATA TYPE VALIDATION

```python
#Print data types of all columns
df.dtypes
```

```python
#removing $ in salary
df['salary'] = df['salary'].str.replace('$','')
df['salary']
```

```python
#Identify columns incorrectly typed as object but should be numeric or datetime
df['salary']=df['salary'].astype(float)
df['salary'].dtypes
```

```python
# Convert the ‘Age’ column to datetime
df['age'] = df['age'].astype('datetime64[ns]')
df['age'].dtypes
```

```python
#Assuming ‘Age’ represents date of birth, compute age_years
df['age_years'] = (pd.to_datetime('today') - df['age']).dt.days / 365
df['age_years'] = df['age_years'].astype(int)
df['age_years']
```

```python
from datetime import date

def calculate_age(born):
    today = date.today()
    # Subtract birth year from current year, then adjust if birthday hasn't passed
    return today.year - born.year - ((today.month, today.day) < (born.month, born.day))
```

```python
df['age_years'] = df['age'].apply(calculate_age)
df['age_years']
```

```python
#Identify invalid date entries and handle them.
#print the value from the column age if the month is more than 12 or les than 0
df[df['age'].dt.month > 12]
```

```python
df[df['age'].dt.month < 0]
```

```python
#print the value from the column age if the day is more than 31 or less than 0 and more than 28 if month is 2
df[(df['age'].dt.day > 28) & (df['age'].dt.month == 2)]
```

```python
df[(df['age'].dt.day > 31) | (df['age'].dt.day < 1)]
```

 SECTION C: SALARY CLEANING

```python
#Convert ‘SALARY’ from currency strings (e.g., $92.32) to float
df['salary'] = df['salary'].astype(float)
df['salary']
```

```python
#Detect missing or malformed salary values
df['salary'].isnull().sum()
```

```python
# Compute minimum, maximum, and average salary after cleaning.
df['salary'].min()
```

```python
df['salary'].max()
```

```python
df['salary'].mean()
```

```python
#Flag salaries equal to zero or negative.
df[df['salary'] <= 0]
```
