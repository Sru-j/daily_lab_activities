```python
#word cloud of the above df on the column name cwe_id
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
```

```python
all_words = ' '.join(df['Status'].str.lower())
```

```python
# Define stopwords (optional, but recommended for better results)
stopwords = set(STOPWORDS)
#stopwords.update(['python', 'data', 'visualization', 'tool']) # Add custom words to ignore

wordcloud = WordCloud(width=800, height=400, background_color='white', stopwords=stopwords, min_font_size=10).generate(all_words)
```

```python
plt.figure(figsize=(10, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off") # Turn off the axis labels
plt.show()
```
