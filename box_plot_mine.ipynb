```python
# Find outlier cereals in sugars using the IQR rule.
Q1 = df["sugars"].quantile(0.25)
Q3 = df["sugars"].quantile(0.75)
IQR = Q3 - Q1

low = Q1 - 1.5 * IQR
high = Q3 + 1.5 * IQR

outliers = df[(df["sugars"] < low) | (df["sugars"] > high)]
print(outliers[["name", "sugars"]] if "name" in df.columns else outliers[["sugars"]])
```

```python
#how to use scatterplot
import matplotlib.pyplot as plt
plt.scatter(df['calories'],df['sugars'])
plt.xlabel('sugars consumed')
plt.ylabel('calories collected')
plt.title('sugars vs. Calories collected')
plt.show()
```

```python
correlation = df['calories'].corr(df['sugars'])
print(f"Correlation: {correlation}")

# Create the scatter plot using pandas' built-in plotting function which uses matplotlib
df.plot(kind='scatter', x='calories', y='sugars', title=f'Scatter Plot (Correlation: {correlation:.2f})')

# Display the plot
plt.show()
```
